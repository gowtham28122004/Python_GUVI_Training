```python
import pandas as pd
import statsmodels.formula.api as smf
```

```python
auto = pd.read_csv(r'D:\Auto.csv')
```

```python
auto.head()
```

regression = smf.ols( data = name stored , formula = dep ~ ind )

```python
# reg = smf.ols( data = auto , formula = dep ~ ind )

model = smf.ols(formula="mpg ~ weight+cylinders+acceleration+displacement+horsepower+year+origin", data = auto)
```

##### Parameter Estimation: 
It computes the values of the model's parameters (e.g., coefficients in linear regression) that minimize the difference between the observed values and the predicted values.
<br>
##### Model Fitting: 
By fitting the model to the data, it learns the relationship between the independent variables (predictors) and the dependent variable (outcome), allowing for predictions on new data.
<br>
##### Return Value: 
The result stored in res typically contains details about the fit, including the coefficients, statistical significance, residuals, and goodness-of-fit measures, which can be accessed using methods like .summary().

```python
res = model.fit()
```

```python
x = ad_data[['Quantity','Discount','Profit']]
y = ad_data.Sales
```

```python
ad_data.columns
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(x)
x
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn import linear_model

regr = linear_model.LinearRegression()
regr.fit(X,y)
```

```python
y_pred = regr.predict(X)
y_pred
```

```python
regr.coef_
```

```python
regr.intercept_
```

```python
print("R square: {}".format(r2_score(y_true=y,y_pred=y_pred)))
```

##### Mean of residual

```python
import numpy as np
residual = y.values - y_pred
mean_residual = np.mean(residual)
print("Mean of Residual {}".format(mean_residual))
```

##### Very close to zero so all are good
